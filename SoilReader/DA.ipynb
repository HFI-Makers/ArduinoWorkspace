```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read file name, must be a csv
df = pd.read_csv("DATA.CSV")
df.columns
```

```python
df.replace(r' ', np.NaN)
df.dropna()
```

```python
x = df.iloc[:, 0]
r1 = df.iloc[:, 1]
r2 = df.iloc[:, 2]
```

```python
plt.figure(figsize=(100, 10), dpi=256, facecolor='w')
plt.scatter(x, r1, color='r')
plt.scatter(x, r2, color='b')
plt.ylabel("Reading", fontdict={'size': 45})
plt.xlabel("Time", fontdict={'size': 45})
plt.tick_params(labelsize=40)
plt.savefig("DATA.png")
plt.show()
```
